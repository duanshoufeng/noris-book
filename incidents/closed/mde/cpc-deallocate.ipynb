# CPC

## This entry cannot be deleted as it is already in use!

![CPC Deallocate](./images/mde_cpc_deallocate.png)

- If there is some history entries for this error booked with this grip map, and you find that the grip map is wrong, then the proper way is to find these entries, and remove the grid infomation for these entries.

```{note}
Don't delete the history entries, delete is not the first option as always.
Since the grid map is not right, so it make sense to remove the grid information from the entries.
```

````sql
SELECT *
FROM [dbo].[CPC_BOOKING]
WHERE PartNumber = 'CT0001A' -- product number
      AND Image_Side_ID IN
          (
              SELECT S.Id
              FROM [dbo].[CPC_ImageSide] s
                  JOIN [dbo].[Translation_Table] t
                      ON s.TranslationDescription = t.Description
              WHERE t.Language_Group_ID = 2
                    AND T.TEXT = 'Top Deck' -- image side name
          )
      AND error_code_id IN
          (
              SELECT e.ERROR_CODE_ID
              FROM [dbo].[ERRORMATRIX] e
                  JOIN [dbo].[Translation_Table] t
                      ON t.Description = e.Translation_Description
              WHERE t.Language_Group_ID = 2
                    AND T.TEXT = 'Crack' -- error name
          );
````

````sql
SELECT *
FROM [dbo].[CPC_BOOKING_DETAIL]
WHERE BOOKING_ID IN
      (
          SELECT BOOKING_ID
          FROM [dbo].[CPC_BOOKING]
          WHERE PartNumber = 'CT0001A'
                AND Image_Side_ID IN
                    (
                        SELECT S.Id
                        FROM [dbo].[CPC_ImageSide] s
                            JOIN [dbo].[Translation_Table] t
                                ON s.TranslationDescription = t.Description
                        WHERE t.Language_Group_ID = 2
                              AND T.TEXT = 'Top Deck'
                    )
                AND error_code_id IN
                    (
                        SELECT e.ERROR_CODE_ID
                        FROM [dbo].[ERRORMATRIX] e
                            JOIN [dbo].[Translation_Table] t
                                ON t.Description = e.Translation_Description
                        WHERE t.Language_Group_ID = 2
                              AND T.TEXT = 'Crack'
                    )
      );
````

- Delete the details infomation for these bookings.

````sql
DELETE FROM [dbo].[CPC_BOOKING_DETAIL]
WHERE BOOKING_ID IN
      (
          SELECT BOOKING_ID
          FROM [dbo].[CPC_BOOKING]
          WHERE PartNumber = 'CT0001A'
                AND Image_Side_ID IN
                    (
                        SELECT S.Id
                        FROM [dbo].[CPC_ImageSide] s
                            JOIN [dbo].[Translation_Table] t
                                ON s.TranslationDescription = t.Description
                        WHERE t.Language_Group_ID = 2
                              AND T.TEXT = 'Top Deck'
                    )
                AND error_code_id IN
                    (
                        SELECT e.ERROR_CODE_ID
                        FROM [dbo].[ERRORMATRIX] e
                            JOIN [dbo].[Translation_Table] t
                                ON t.Description = e.Translation_Description
                        WHERE t.Language_Group_ID = 2
                              AND T.TEXT = 'Crack'
                    )
      );
````

```{admonition} Tip
:class: tip
We have to delete the grid map information, it is very dangerous but we have no other choise.

The booking entries are still there.
```
